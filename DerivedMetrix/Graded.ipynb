```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
df= pd.read_csv("grades.csv")
df.head()
```

```python
df['is_zip'] = df.submission.apply(lambda x: 1 if 'zip' in x.split('.')[-1] else 0)
```

```python
###Jan 3, 2017 - 11:59:59
###Jan 9, 2017 - 11:59 PM
##df.head(1).submit_time.astype('datetime64[ns]')
from datetime import datetime
from dateutil import parser
first_deadline= parser.parse("Jan 3, 2017 - 11:59:59PM")
second_deadline= parser.parse("Jan 9, 2017 - 11:59:59PM")
first_deadline > second_deadline
```

```python
df.submit_time = df.submit_time.astype('datetime64[ns]')
```

```python
def submission_after1(x):
    return 1 if x> first_deadline and x <second_deadline else 0
```

```python
def submission_after2(x):
    return 1 if x> second_deadline else 0
```

```python
df['after_1'] = df.submit_time.apply( lambda x:submission_after1(x) )
df['after_2'] = df.submit_time.apply( lambda x:submission_after2(x) )
```

```python
df
```

```python
df[ (df.after_1 == 1) | (df.after_2 == 1) ].count()
```

```python
df['date'] = df.submit_time.apply(lambda x: pd.to_datetime(x).date() )
```

```python
df.groupby('date').count()
```

```python
df['hour'] = df.submit_time.apply(lambda x: pd.to_datetime(x).time().hour )
```

```python
df.groupby('hour')['hour'].count()
```
