```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("tendulkar_ODI.csv")
```

```python
df.Runs.isnull().sum()
```

```python
df= df[~df.Runs.isin(['DNB','TDNB'])]
```

```python
df = df[df['4s'] !='-']
```

```python
df.Runs = df.Runs.dropna(how='any', inplace=True)
```

```python
def clearRun(x):
    print(x)
    if ('*' in x):
        return int(x[:-1])
    return int(x)
```

```python
df.Runs = df.Runs.apply(lambda x: clearRun(x))
```
